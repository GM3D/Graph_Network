## やや冗長に述べた定理 2.2(p.28)の証明

### preorderの場合

preorderに関しては、新しい頂点を見つけるごとに即座にラベルをつけていくので、一つの木の中で子孫が親より大きいラベルを持つのは当たり前であり、前進辺、後退辺もそれぞれその始点と終点が探索木内の先祖と子孫の関係にある以上、それぞれの向きにより(a)、(b)が成り立つのは明らか。自明でないのは(c)の横断辺の場合なので、この場合について考える。背理法で横断辺(u, v)に対してprelabel(u) < prelabel(v)であったとする（ラベルは頂点ごとに異なるので等号はありえない）。これは、ちょうど頂点uを起点に辺(u, v)が探索されている時点を考えると、preorderを仮定しているのでラベルの大小を考えれば、まだvはその時点では未探索であったことになる。するとuからvへ向かう未探索（今まさに探索しようとしている）辺がありその先が未探索なのだから、深さ優先アルゴリズムに従う限り、(u, v)がここで探索済みとされ、この時点でvに新しくラベル付けがされることになる。ところがこれはまさに(u, v)を木辺としたことに他ならないから(u, v)が横断辺という仮定に反する。したがって prelabel(u) > prelabel(v)でなければならない。

### postorderの場合

postorderの場合は、探索木の根からある末端ノードまでの一本のパスに注目すれば、その末端から引き返しながらラベルをつけていくのでそのパスの中では末端から根までの順にラベルは単調に増加（途中他の枝の探索が入るとそこでラベルに飛びができるが、増加することに変わりはない）、逆に根から末端の向きで見れば単調減少。よって(a)、(b)に関してはprelabelのときの逆順になる。
(c)、横断辺の場合は、その横断辺を(u, v)とし、この辺を探索する時点を考えてみる。この辺が横断辺であって木辺ではないのだから、この辺の先にある頂点vはすでに探索済みであったということであり、しかも今は頂点uが属している枝を探索中なのだから、vの属している枝については、少なくとも先端からvまでの部分については探索が完了しておりすでに後行順でもラベル付けが済んでいることになる。それに対してuは今現在か、もしくはuから出る未探索辺がまだ他にも存在していればもっと後にラベル付けされることになるので、postlabel(v) < postlabel(u)となる。

## 定理3.1 (p. 207) の説明の前段部分（実質的に独立した命題になっている）について

Gを有向グラフ、v, wをその任意の相異なる二頂点とする。

命題. vとwが同じ強連結成分に属するなら、深さ探索で同じ根付き木に属する。

証明: 強連結成分の定義から、有向パスP1(v, w)とP2(w, v)がGに存在する。vが属する根付き木をA、wが属する根付き木（Aと異なる）をBとすると、根付き木がAとBしかない場合（一般の場合は最後に議論する）、P1はAからBへ移る辺（したがって横断辺）を含み、同じくP2はBからAへ移る横断辺を含む。

Aの根をr、Bの根をr'とし、P1のAからBへの横断編を(s, t)とする。すると、定理2.2よりprelabel(s) > prelabel(t),またtはr'の子孫だからprelabel(t) > prelabel(r')ここで、この深さ優先探索で初めてsが探索された時点のことを考えると、この時点でsより若いprelabelを持つtやr'を含む木Bはすでにすべて探索済みであったことがわかる。それに対して今まさにsが探索されようとしているので、sを含む木AはBより後に探索されているということになる。したがって、AとBの根であるrとr'についても、r'の方が先に探索されており、prelabel(r') < prelabel(r)でなければならない。ところがまったく同じ議論を辺P2に適用することによりprelabel(r) < prelabel(r')であることもいえ、これは矛盾である。よってP1とP2はそのような異なる根付き木を結ぶ辺を持たない。すなわちv、wは同じ根付き木に属する。

一般の場合、AからBへ行く途中でP1やP2(一方、あるいは両方）が別の根付き木Cをさらに経由している可能性もあるが、上の議論と同じく別の木に移るときに必ず木の根のprelabelは減少するのでvとwの属する木の根rとr'に関してはやはり上の両方の不等式が成り立つことになり、A=Bでないかぎり矛盾となる。

ただしこの命題１は定理3.1の証明に直接は使われない。p. 207では、この命題の逆は成り立たないため、根付き木の分類だけでは強連結成分分解はできない、ということの説明に使われているのみである。定理3.1自体の証明はp. 207に述べられている通りで、その説明は十分詳細なのでここであらためて述べることはしない。

## 補題6.1 (p. 84) の証明 (演習問題6.1 (p. 97)、解答 p. 209)についての細かいメモ

p. 209に述べられていることについて、基本的に蛇足だが後で見返すときのため細かい注を加えていく。
まず、証明したいことは 「$G$に$M$に関する増加パスが存在しない$\Leftrightarrow$ $M$は最大マッチング」。一方で本文で説明されているように、「$G$に$M$に関する増加パスが存在する$\Rightarrow$（そのパスの$M$に含まれる辺と含まれない辺を反転してより大きなマッチングができるので）$M$は最大マッチングではない」ことがすぐわかるので、これの対偶を取って「$M$が最大マッチング$\Rightarrow$ $G$に$M$に関する増加パスが存在しない」は言える。つまり証明したいことのうち（$\Leftarrow$）は言えているので、逆向きの「$G$に$M$に関する増加パスが存在しない$\Rightarrow$$ $M$は最大マッチング」を言えばいい。グラフ$G$に対して、最大マッチングは複数あり得るので、$M$が最大マッチングであるということは(最大とは$G$に関するマッチングの中で要素数、つまり辺の数が最大であるということに注意すると)、他の任意の最大マッチング$M^\ast$と同じ数の要素を持つ、つまり一般に集合$S$の要素数を$|S|$という記号で書くことにすれば、$|M|=|M^\ast|$といってもよい。

そこで、$M$は$G$のマッチングで増加パスを持たないものとし、$M^\ast$をある最大マッチングとして$|M|=|M^\ast|$を示すという戦略をとる。

さらに$|M|=|M^\ast|$を示すために、テキストの以下の図を考えると、$M = (M - M^\ast) \cup (M \cap M^\ast)$ である。要素数を考えると、$M - M^\ast$ と $M \cap M^\ast$ は共通部分がないので、単に両者の要素数を足して $|M| = |(M - M^\ast)| + |(M \cup M^\ast)|$。 同様に$M^\ast$の要素数を考えれば$|M^\ast| = |(M^\ast - M)| + |(M \cup M^\ast)|$。したがって$|M|=|M^\ast|$を示すには$|(M - M^\ast)|=|(M^\ast - M)|$を示せばよい。以下これを目標にする。

![$M$と$M^\ast$の関係](fig1.png)

上図の網掛け部分全体、$(M - M^\ast) \cup (M^\ast - M)$を以下$M\Delta M^\ast$と表記する。$M\Delta M^\ast$から誘導される$G$の部分グラフ（$M$と$M^\ast$は$G$の辺の集まりなので、それらの辺と、その辺の両端の頂点からなるグラフ）$G|(M\Delta M^\ast)$(長いので以下$G^\prime$とする）を考えると、$E(G') = M\Delta M^\ast$ であって、$G^\prime$の辺全体は$M\Delta M^\ast$と一致しており、$(M - M^\ast) \cup (M^\ast - M)=M\Delta M^\ast$において、「$M$の辺」といえば$M - M^\ast$の辺というのと同じであり、「$M^\ast$の辺」といえば$M^\ast - M$の辺と同じであることに注意する。

今$G^\prime$の任意の頂点$v$を考えると、$G^\prime$の作り方からこれは$M$に含まれる辺かもしくは$M^\ast$に含まれる辺のいずれかの端点になっている（両方の端点になっているかもしれない）。$M$は$G$のマッチングであることから、$M$に含まれる2つの辺が同じ頂点を共有することはない。$M^\ast$についても同じ。ということは頂点$v$に接続されている$M$の辺はたかだか一本、$M^\ast$についてもたかだか一本。両方合わせて高々2本。（なお両方合わせて0本ということはない。なぜなら0本であったならその頂点はもともと$M$の辺とも$M^\ast$の辺とも繋がっていないということであり、誘導グラフに含まれない）
つまり、$G^\prime$というグラフはその全頂点が必ず一本または２本の辺に接続されている。$G^\prime$は一般にいくつかの連結成分からなっているだろうが、このことからその一つの連結成分を考えると、それは分岐できない（分岐点は３つ以上の辺に接続されているので）ため、単純なパス$P$か閉路$C$かのいずれかになる。$G^\prime$全体は、いくつかのパスと閉路を合わせたものということになる。
この$G^\prime$の中で、$M$に含まれる辺と$M^\ast$に含まれる辺の数を数えていき両者が等しいことを示せばよい。なお$G^\prime$は$M$と$M^\ast$の共通部分は取り除いた$M\Delta M^\ast$を基にしているので、$G^\prime$の辺は必ず$M$に含まれるか$M^\ast$に含まれるかのいずれかであって、両方に含まれることはないことに注意する。

まず$C$から考えると、$M$も$M^\ast$もマッチングである以上一つの頂点を二つの$M$の辺（あるいは二つの$M^\ast$の辺）が共有することはないのだから、各頂点に$M$の辺と$M^\ast$の辺が一本ずつ接続されるしかない。つまり、$M$の辺と$M^\ast$の辺が必ず交互に現れるループになっており、したがって$M$と$M^\ast$の辺が同じ個数現れ、辺と頂点をそれぞれ偶数個もつようなループになっている。したがって$G^\prime$のうちループを構成する部分については、$M$と$M^\ast$の辺が同数現れることが分かった。

![閉路$P$の場合](fig2.png)

次に単純パス$P$の場合を考えると、やはり$M$および$M^\ast$がマッチングである以上、ループのときと同じ理由により$M$の辺と$M^\ast$の辺が交互に現れる。$G^\prime$が$G$の誘導部分グラフなのだから$G^\prime$のパスは$G$のパスでもある。$G$は今仮定により$M$に関する増加パスを持たないので、$G^\prime$も$M$に関する増加パスを持ってはならない。$M$に関する増加パス、とはなんであったかというと、$M$の辺と$M$以外（つまり今の場合必然的に$M^\ast$)の辺が交互に現れるパスで、最初の辺も最後の辺も$M$に含まれない（つまり$M^\ast$の辺で始まり$M^\ast$の辺で終わる）ものである。そういうパスがないといっているのだから、Pは最初と最後少なくとも一方（あるいは両方）が$M$の辺である。ということは、Pに含まれる辺のうち$M$に含まれるものの数$n(P, M)$と$M^\ast$に含まれるものの数$n(P, M^\ast)$を考えると$n(P, M) \ge n(P, M^\ast)$ということになる。一方で、$M^\ast$は仮定により$G$のある最大マッチングであり、この証明の最初で「簡単に分かる方」として考えたことにより、$G$には、したがってその部分グラフである$G^\prime$には$M^\ast$に関する増加パスは存在しない。したがって、$M$について考えたときと同じく$n(P, M^\ast) \ge n(P, M)$が成り立つ。この二つの逆向きの不等式から、許されるのは $n(P, M^\ast) = n(P, M)$ のみということになる。つまり$G^\prime$の連結成分となっている任意のパスは、$M$の辺と$M^\ast$の辺を（互い違いに）ちょうど同じだけ含む。

![パス$P$の場合](fig3.png)

つまり、まとめると$M\Delta M^\ast$の辺をすべて持ってきて作ったグラフ$G^\prime$はループや単純パスの集まりだが、そのループも単純パスもいずれも$M$と$M^\ast$の辺を同数含むことになる。つまり、『$(M - M^\ast) \cup (M^\ast - M)=M\Delta M^\ast$において、「$M$の辺」といえばM - $M^\ast$の辺というのと同じであり、「$M^\ast$の辺」といえば$M^\ast - M$の辺と同じであることに注意する。』を思い出すと、$|(M - M^\ast)|=|(M^\ast - M)|$が示されたことになり、証明が完了する。
